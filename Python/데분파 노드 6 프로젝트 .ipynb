```python
import sklearn
from sklearn.datasets import load_digits,load_wine, load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression
```

```python
#digits 분류 - Decision Tree, Random Forest 사용
```

```python
digits=load_digits()
print(dir(digits))
```

```python
digits_data=digits.data
digits_label=digits.target
print(digits.target_names)

import pandas as pd
pd_data=pd.DataFrame(digits_data)
pd_data.describe()
```

```python
X_train, X_test, y_train, y_test= train_test_split(digits_data, digits_label, test_size=0.2, random_state=7)
```

```python
decision_tree=DecisionTreeClassifier(random_state=32)
decision_tree.fit(X_train, y_train)
y_pred=decision_tree.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
random_forest = RandomForestClassifier(random_state=32) 
random_forest.fit(X_train, y_train) 
y_pred = random_forest.predict(X_test) 

print(classification_report(y_test, y_pred))
```

```python
#정밀도:precision으로! 틀렸는데 맞았다고 하는게 더 안 좋은 것 같음
```

```python
## wine 분류 (class 0/1/2) -SVM, SGD 사용
```

```python
wine=load_wine()
dir(wine)
```

```python
wine_data=wine.data
wine_label=wine.target
print(wine.target_names)

import pandas as pd
pd_data=pd.DataFrame(wine_data)
pd_data.describe()
```

```python
X_train, X_test, y_train, y_test= train_test_split(wine_data, wine_label, test_size=0.2, random_state=7)
```

```python
svm_model=svm.SVC()

svm_model.fit(X_train, y_train)
y_pred=svm_model.predict(X_test)

print(classification_report(y_test,y_pred))
```

```python
sgd_model=SGDClassifier()

sgd_model.fit(X_train, y_train)
y_pred=sgd_model.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
#f1-score: 0등급, 1등급, 2등급 골고루 된 것 같아서
```
